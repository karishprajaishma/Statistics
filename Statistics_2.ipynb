```python
#IMPORTING NECESSARY PACKAGES
import numpy as np
import scipy
import matplotlib.pyplot as plt

#GENERATING X VALUES
x=np.arange(-10,10,0.1)

#FINDING PDF FOR EACH X VALUE FOR STD.NORMAL DISTRIBUTION
y=scipy.stats.norm.pdf(x,loc=0,scale=1)

#PLOTTING X AND PDF
plt.plot(x,y)

#DOTTED LINES FOR MEAN
plt.axvline(0,ls="--",c="lightblue")

```

```python
##FINDING CDF FOR EACH X VALUE FOR STD.NORMAL DISTRIBUTION
y=scipy.stats.norm.cdf(x,loc=0,scale=1)

#PLOTING X AND CDF
plt.plot(x,y)

#FINDING 1 STD,DEVIATION AREA
c= scipy.stats.norm.cdf(1,0,1)-scipy.stats.norm.cdf(-1,0,1)

#PLOTTING THE AREA
plt.fill_between(x, y, where=(x >= -1) & (x <= 1), alpha=0.2, color='blue')

#PRINTING VALUE
print(c)
```

```python
#CALCULATING THE CDF OF THE MID VALUE 
a=scipy.stats.norm.cdf(0,0,1)
#PRINT CDF (HALF AREA)
print(a)

#CALCULATING PPF AT 0.5
b=scipy.stats.norm.ppf(0.5)
print(b)

#FINDING 1 STD,DEVIATION AREA AND PRINTING
c=scipy.stats.norm.cdf(1,0,1)-scipy.stats.norm.cdf(-1,0,1)
print(c)

#FINDING 2 STD,DEVIATION AREA AND PRINTING
d=scipy.stats.norm.cdf(2,0,1)-scipy.stats.norm.cdf(-2,0,1)  
print(d)

#FINDING 3 STD,DEVIATION AREA AND PRINTING
e=scipy.stats.norm.cdf(3,0,1)-scipy.stats.norm.cdf(-3,0,1)
print(e)
```

```python
#X FROM VALUE 0 TO 6 , CONSIDERING 500 VALUES
x=np.linspace(0,6,500)

#FINDING PDF FOR X WHEN ASSUMING MEAN AS 3 AND STD=[1,2,3]
y=scipy.stats.norm(loc=3,scale=1).pdf(x)
y1=scipy.stats.norm(loc=3,scale=2).pdf(x)
y2=scipy.stats.norm(loc=3,scale=3).pdf(x)

#PLOTTING X AND PDF
plt.plot(x,y )
plt.plot(x,y1,c='r')
plt.plot(x,y2,c='g')

#PLOTTING X AND CDF
plt.plot(x,scipy.stats.norm.cdf(x,3,1))
plt.plot(x,scipy.stats.norm.cdf(x,3,2),c='r')
plt.plot(x,scipy.stats.norm.cdf(x,3,3),c='g')

#LIGHTBLUE DASHED LINE FOR MEAN
plt.axvline(3,ls="--",c="lightblue")
```

```python
#GENERATING RANDOM NUMBERS DATA FROM NORMAL DISTRIBUTION
data = scipy.stats.norm.rvs(size=10000)

#PLOTTING HISTOGRAM
plt.hist(data, bins=50, density=True)

#VERIFICATION OF NORMAL DISTRIBUTION USING PLOTTING DATA WITH RESPECT TO PDF
plt.plot(data,scipy.stats.norm.pdf(data) ,alpha=0.3)


```

```python
#ASSUMING UNIFROM DISTRIBUTION IN RANGE OF 0 TO 10
a,b=0,10

#GENERATING 100 VALUES BETWEEN 0 TO 10
x = np.linspace(a, b, 100)

#CALCULATING PDF
pdf = scipy.stats.uniform.pdf(x, loc=a, scale=b)

#CALCULATING CDF 
cdf = scipy.stats.uniform.cdf(x, loc=a, scale=b)

#PLOTTING PDF
plt.plot(x, pdf)

#PLOTTING CDF
plt.plot(x, cdf,c='r')
```

```python
#ASSUMING UNIFORM DISTRIBUTION OF DIFEFRENT INTERVALS I1[0-10] ,I2[2-11], I3[3,12]
a=[0,2,3]
b=[10,11,12]

#GENERATING 100 VALUES BETWEEN 0 TO 10
x = np.linspace(a, b, 100)

#CALCULATING PDF
pdf = scipy.stats.uniform.pdf(x, loc=a, scale=b)

#CALCULATING CDF
cdf = scipy.stats.uniform.cdf(x, loc=a, scale=b)

#PLOTTING PDF
plt.plot(x, pdf)

#PLOTTING CDF
plt.plot(x, cdf,c='r')
```

```python
#GENERATING 1000 VALUES FROM UNIFORM DISTRIBUTION ,RANGE OF 0 TO 1 
x_u = np.random.uniform(0, 1, 1000)

#CONVERTING TO NORMAL DISTRIBUTION USING INVERSE TANSFORMATION IE)CONSIDERING PPF
x_n = scipy.stats.norm.ppf(x_u) 

#PLOTTING UNIFROM DISTRIBUTION HISTOGRAM
plt.hist(x_u, bins=10, alpha=0.7)

#PLOTTING NORMAL DISTRIBUTION HISTOGRAM
plt.hist(x_n, bins=10,alpha=0.3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

#GENERATING 1000 VALUES FROM NORMAL DISTRIBUTION ,RANGE OF 0 TO 1 
x_n =np.random.normal(0,1,1000)

#CONVERTING TO UNIFORM DISTRIBUTION USING INVERSE TANSFORMATION IE)CONSIDERING CDF
x_u=scipy.stats.norm.cdf(x_n)


#PLOTTING NORMAL DISTRIBUTION HISTOGRAM WITH TRANSPARENCY LEVEL OF 0.5
plt.hist(x_n,alpha=0.5)

#PLOTTING UNIFROM DISTRIBUTION HISTOGRAM
plt.hist(x_u)

```
