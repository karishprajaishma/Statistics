```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1) y= e**(-x)

#ARRAY IN THE RANGE OF -10 TO +10 WITH THE STEP INCREASE OF 0.1
x = np.arange(-10,10.1,0.1)

#PASSING X IN THE FUNCTION AND GETTING VALUES FOR Y 
y = np.exp((-x))
print(y)

#PLOTTING X, Y
plt.plot(x,y)
```

```python
#2) y= e**|-x|

#ARRAY IN THE RANGE OF -10 TO +10 WITH THE STEP INCREASE OF 0.1
x = np.arange(-10,10,0.1)

#PASSING X IN THE FUNCTION AND GETTING VALUES FOR Y 
y = np.exp(-abs(x))

#PLOTTING X, Y
plt.plot(x,y)
```

```python
#3) y=e**(-(x**2)/2)

#ARRAY IN THE RANGE OF -10 TO +10 WITH THE STEP INCREASE OF 0.1
x = np.arange(-10,10,0.1)

#PASSING X IN THE FUNCTION AND GETTING VALUES FOR Y 
y=np.exp(-(x**2)/2)

#PLOTTING X, Y
plt.plot(x,y)
```

```python
# 4) y=e(-(x**2+y**2)/2)

# X AND Y ARE  INDEPENDENT VARIABLES UNLIKE PREVIOUS SINGLE INDEPENDENT VARIABLE

#INTIALIZING 3D PROJECTION AXES
ax=plt.axes(projection='3d')

#ASSIGN X,Y VARIABLES WITH ARRAY IN THE RANGE OF -10 TO +10 WITH THE STEP INCREASE OF 0.1
x = np.arange(-10,10,0.1)
y = np.arange(-10,10,0.1)

#PASSING X AND Y IN THE FUNCTION AND GETTING VALUES FOR Z
z = np.exp(-((x**2+y**2)/2))

#PLOTTING
ax.plot(x,y,z)
plt.show()
```

```python
#5) INTEGRATE AND FIND THE VALUE FOR GAUSSIAN FUNCTION y=e**(-(x**2)/2)


import scipy
from scipy.integrate import quad

#INTEGRATION FOR GAUSSIAN FUN
def integrand(x):
    return np.exp(-(x**2)/2)

#RETURN RESULT AND ERROR 
result, error = quad(integrand, -np.inf, np.inf)

print(result)
```

```python
#7) f(x) = e**x/1+e**x


#ARRAY IN THE RANGE OF -10 TO +10 WITH THE STEP INCREASE OF 0.1
x = np.arange(-10,10,0.1)

#PASSING X IN THE FUNCTION AND GETTING VALUES FOR Y 
y = np.exp(x)/1+np.exp(x)

#PLOTTING
plt.plot(x,y)
```

```python
#IMPORTING PYTHON PACKAGE SCIPY
import scipy

#18 VALUES OF HEIGHTS DATA 
x=np.array([5.0,5.1,5.2,5.3,5.4,5.5,5.5,5.6,5.6,5.6,5.6,5.7,5.7,5.8,5.8,5.9,6,6.2])

#CALCULATING MEAN
m=np.mean(x)

#CALCULATING STD. DEVIATION
s=np.std(x)

#PRINTING MEAN ,STD DEVIATION AND HEIGHT DATA
print(m)
print(s)
print(x)

#PASS THE MEAN AND STD DEVIATION ALREADY CALCULATED IN PDF FUNCTION
plt.plot(x, scipy.stats.norm.pdf(x, m, s))

#PLOT HISTOGRAM OF BIN SIZE =15 TO ACCESS THE DISTRIBUTION
plt.hist(x,bins=15)

#PLOTTING CDF
plt.plot(x, scipy.stats.norm.cdf(x,m,s),c='r')

```

```python
#INPUT- NO OF HEADS MAX POSSIBLE (0 TO 4) WITH STEP OF 1
x = np.arange(0,4,0.1)  

#NO OF COINS TOSSES 
n = 3 

#PROBABILITY OF GETTING EITHER HEAD/TAIL IN EACH TOSS
p = 0.5 

#CALCULATING PMF
noheads = scipy.stats.binom.pmf(x, n, p)

#CALCULATING CDF
cdf=scipy.stats.binom.cdf(x,n,p)

#PLOTTING X AND PMF
plt.plot(x,noheads)

#PLOTTING X AND CDF
plt.plot(x,cdf)
```

```python
#X-VALUES FROM -1 TO 1
x1=np.arange(-1,0.1,0.1)

#EQN FOR LINE FOR POINTS(-1,0) AND (0,1)
y1=x1+1

#X-VALUES FROM 0 TO 1
x2=np.arange(0,1.1,0.01)

#EQN FOR LINE FOR POINTS(0,1) AND (1,0)
y2=-x2+1

#CONCATENATING X1 AND X2 IE) JOINING BOTH LINES TO FORM TRIANGLE
x=np.concatenate([x1,x2])

#CONCATENATING Y1 AND Y2 IE) JOINING BOTH Y GENERATED VALUES
y=np.concatenate([y1,y2])

# print(x)
# print(y)

#HEIGHT IS O.5
c=0.5

#INTIALIZE LIST FOR CDF
cd=[]

#PLOTTING X AND Y (PDF)
plt.plot(x,y)

#A AND B ARE THE RANGE OF START AND END VALUE OF BASE OF THE TRIANGLE 
a = -1  
b = 1   

def cdf(x, c, a, b):
    #IF THE X IS LESS THAN 0,THEN NO AREA ,THUS RETURN 0
    if x < a:
        return 0
    
    #IF X IN RANGE OF A TO C ,THEN AREA UNDER A TO C IS CALCULATED BY FOLLOWING FORMULA
    elif a <= x < c:
        return (x - a)**2 / ((b - a) * (c - a))
    
    #IF X IN RANGE OF C TO B ,THEN AREA UNDER C TO B IS CALCULATED BY FOLLOWING FORMULA
    elif c <= x <= b:
        return 1 - (b - x)**2 / ((b - a) * (b - c))
    #IF X BEYOND B ,THEN AREA IS ONE
    else:
        return 1

#PASSING (X_VAL)-INDIVIDUAL VALUES IN FUNCTION AND CALCULATING CDF FOR EACH DIFFERENT CASE
cd= [cdf(x_val, c, a, b) for x_val in x]

#PLOTTING X AND CDF
plt.plot(x, cd,c='r')
```
